## 合并拆分操作

```python
d1 = np.random.normal(1,1,size=(10))
d2 = d1 * 2
```

```python
d1
```

```python
d2
```

```python
# 将两个1维数组合并, axis 按行或列进行拼接
# concatenate 只适用于维度匹配的矩阵
z = np.concatenate([d1,d2])
w = np.concatenate([[z,z],[z * 2, z*2]], axis=1)
```

```python
# np.vstack/hstack 容错性更好的 垂直合并
np.hstack([z,np.zeros(shape=(20))])
```

```python
# np.vsplit/h.split : 在垂直和水平进行分割
c1,c2,others = np.hsplit(w,[1,2])
others
```

## 矩阵运算

```python
# Uninversal Function
n = 10
L = [_ for _ in range(n)]
A = np.array(L)
np.log(A+1)
```

```python
# 矩阵乘法
a = np.full((2,2),1)
b = np.full((2,2),12)
a.dot(b)
```

```python
# 向量和矩阵运算
v = np.array([1,2])
a + np.vstack([v] * a.shape[0])
```

```python
# 矩阵的逆 - 方阵可求逆
a = np.random.normal(1,1,(5,5))
a.dot(np.linalg.inv(a))
```

## Numpy 索引

```python
# argument 返回固定数值的索引

import numpy as np
data = np.random.normal(0,1000,size=500)
data.argmin()
```

```python
# 排序和使用索引, arg, arg
np.random.shuffle(data.reshape((5,-1)))
np.sort(data, axis=0)
print(data)

# argsort
np.argsort(data)

np.partition(data,0)
```
